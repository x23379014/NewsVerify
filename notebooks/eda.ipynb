# Exploratory Data Analysis - Fake News Detection Dataset

This notebook performs EDA on the fake news detection dataset.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load data
df = pd.read_csv('../data.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Basic info
print("Dataset Info:")
print(df.info())
print("\nMissing values:")
print(df.isnull().sum())
print("\nLabel distribution:")
print(df['Label'].value_counts())
```

```python
# Visualize label distribution
plt.figure(figsize=(8, 6))
df['Label'].value_counts().plot(kind='bar', color=['#667eea', '#764ba2'])
plt.title('Label Distribution')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
